### Valg av datasett

I denne oppgaven valgte jeg å bruke Tingvold large (54)-datasettet. Dette datasettet dekker et større område (31 daa) og består av omtrent 2500 bilder tatt med 40° gimbalvinkel. Jeg valgte dette fordi jeg ønsket en mer realistisk og utfordrende case der både datamengden og bildenes variasjon setter krav til preprocessing og modelltrening.
Skrå vinkler og varierende lysforhold gjør det mer krevende å annotere bildene, men samtidig gir det et mer robust grunnlag for å utvikle en modell som generaliserer godt til reelle droneforhold. Den store datamengden åpner også for å teste hvordan ulike preprocessing- og augmenteringsstrategier påvirker ytelsen.
